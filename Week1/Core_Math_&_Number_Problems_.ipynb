```python
def sum_of_digits(n):
    n = abs(n)          # handle negative numbers
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

n = int(input("Enter an integer: "))
print("Sum of digits:", sum_of_digits(n))
```

```python
def product_of_digits(n):
    n = abs(n)          # handle negative numbers
    product = 1

    if n == 0:
        return 0

    while n > 0:
        product *= n % 10
        n //= 10

    return product


n = int(input("Enter an integer: "))
print("Product of digits:", product_of_digits(n))
```

```python
def reverse_number(n):
    sign = -1 if n < 0 else 1
    n = abs(n)

    rev = 0
    while n > 0:
        rev = rev * 10 + (n % 10)
        n //= 10

    return sign * rev


n = int(input("Enter a number: "))
print("Reversed number:", reverse_number(n))
```

```python
def is_palindrome(n):
    if n < 0:
        return False    # negative numbers are not palindrome

    original = n
    rev = 0

    while n > 0:
        rev = rev * 10 + (n % 10)
        n //= 10

    return original == rev


n = int(input("Enter a number: "))
if is_palindrome(n):
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

```python
def is_armstrong(num):
    if num < 0:
        return False

    # Count digits
    n = 0
    temp = num
    while temp > 0:
        n += 1
        temp //= 10

    # Calculate Armstrong sum
    temp = num
    armstrong_sum = 0
    while temp > 0:
        digit = temp % 10
        armstrong_sum += digit ** n
        temp //= 10

    return armstrong_sum == num


num = int(input("Enter a number: "))

if is_armstrong(num):
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

```python
def is_perfect(num):
    if num <= 1:
        return False

    divisor_sum = 1  # 1 is always a proper divisor
    i = 2

    while i * i <= num:
        if num % i == 0:
            divisor_sum += i
            if i != num // i:
                divisor_sum += num // i
        i += 1

    return divisor_sum == num


num = int(input("Enter a number: "))

if is_perfect(num):
    print("Perfect number")
else:
    print("Not a Perfect number")
```

```python
def is_strong(num):
    if num < 0:
        return False

    original = num
    strong_sum = 0

    while num > 0:
        digit = num % 10

        # factorial of digit
        fact = 1
        for i in range(1, digit + 1):
            fact *= i

        strong_sum += fact
        num //= 10

    return strong_sum == original


num = int(input("Enter a number: "))

if is_strong(num):
    print("Strong number")
else:
    print("Not a Strong number")
```

```python
def is_harshad(num):
    if num <= 0:
        return False

    temp = num
    digit_sum = 0

    while temp > 0:
        digit_sum += temp % 10
        temp //= 10

    return num % digit_sum == 0


num = int(input("Enter a number: "))

if is_harshad(num):
    print("Harshad (Niven) number")
else:
    print("Not a Harshad number")
```

```python
def is_automorphic(num):
    if num < 0:
        return False

    square = num * num

    temp = num
    while temp > 0:
        if temp % 10 != square % 10:
            return False
        temp //= 10
        square //= 10

    return True


num = int(input("Enter a number: "))

if is_automorphic(num):
    print("Automorphic number")
else:
    print("Not an Automorphic number")
```

```python
def is_neon(num):
    if num < 0:
        return False

    square = num * num
    digit_sum = 0

    while square > 0:
        digit_sum += square % 10
        square //= 10

    return digit_sum == num


num = int(input("Enter a number: "))

if is_neon(num):
    print("Neon number")
else:
    print("Not a Neon number")
```

```python
n = int(input("Enter the upper limit: "))

# Step 1: Assume all numbers are prime
is_prime = [True] * (n + 1)
is_prime[0] = is_prime[1] = False

# Step 2: Mark multiples as non-prime
i = 2
while i * i <= n:
    if is_prime[i]:
        for j in range(i * i, n + 1, i):
            is_prime[j] = False
    i += 1

# Step 3: Print primes
print("Prime numbers are:")
for i in range(2, n + 1):
    if is_prime[i]:
        print(i, end=" ")
```

```python
n = int(input("Enter the upper limit: "))

# Step 1: Sieve of Eratosthenes
is_prime = [True] * (n + 1)
is_prime[0] = is_prime[1] = False

for i in range(2, int(n**0.5) + 1):
    if is_prime[i]:
        for j in range(i * i, n + 1, i):
            is_prime[j] = False

# Step 2: Find twin primes
print("Twin primes are:")
for i in range(3, n - 1):
    if is_prime[i] and is_prime[i + 2]:
        print(f"({i}, {i + 2})", end=" ")
```

```python
def factorial(n):
    if n == 0 or n == 1:   # base case
        return 1
    return n * factorial(n - 1)

n = int(input("Enter a number: "))
print("Factorial:", factorial(n))
```

```python
def factorial(n):
    res = 1
    for i in range(2, n + 1):
        res *= i
    return res

n = int(input("Enter n: "))
r = int(input("Enter r: "))

# nCr is not defined if r > n
if r > n:
    print("Invalid input")
else:
    r = min(r, n - r)   # simplification
    ncr = factorial(n) // (factorial(r) * factorial(n - r))
    print("nCr =", ncr)
```

```python
import math

x = float(input("Enter x (in degrees): "))
n = int(input("Enter number of terms: "))

# Convert degrees to radians
x = math.radians(x)

sinx = 0
sign = 1
power = x
fact = 1

for i in range(1, 2*n, 2):
    sinx += sign * power / fact
    sign *= -1
    power *= x * x
    fact *= (i + 1) * (i + 2)

print("sin(x) =", sinx)
```
